# Linear Regression on Medical Price DataSet

```python
# loading the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# performing the linear regression making it as a function
def linear_regression(x_train, y_train):
    # Add a column of ones to the feature matrix for the bias term
    x_train = np.column_stack((np.ones(len(x_train)), x_train))

    # Calculate the weights using the normal equation
    weights = np.linalg.inv(x_train.T @ x_train) @ (x_train.T @ y_train)

    return weights
```

```python
# Load the medical price dataset
data = pd.read_csv('Medical Price Dataset.csv')
```

```python
# Extract the age and charges from the dataset
x_train = data['age'].values
y_train = data['charges'].values
```

```python
# Apply linear regression
weights = linear_regression(x_train, y_train)
```

```python
# Predict the output using the learned weights
x_test = np.linspace(np.min(x_train), np.max(x_train), 100)
x_test_with_bias = np.column_stack((np.ones(len(x_test)), x_test))
y_pred = x_test_with_bias @ weights
```

```python
# Plot the training data and the learned regression line
plt.scatter(x_train, y_train, label='Training Data',color='green')
plt.plot(x_test, y_pred, color='red', label='Linear Regression')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Linear Regression on medical Price Dataset')
plt.legend()
plt.show()
```
